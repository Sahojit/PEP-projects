```python
# Activity 1: Student Management System using Hash Table (Dictionary)
# Best Time Complexity: O(1) for Insert, Search, Delete

students = {}   # Hash table: roll_no -> student details


def insert_student():
    roll_no = input("Enter Roll Number: ")
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    course = input("Enter Course: ")

    students[roll_no] = {
        "name": name,
        "age": age,
        "course": course
    }

    print("✅ Student inserted successfully!")


def search_student():
    roll_no = input("Enter Roll Number to search: ")

    if roll_no in students:
        s = students[roll_no]
        print("\n--- Student Found ---")
        print("Roll No:", roll_no)
        print("Name:", s["name"])
        print("Age:", s["age"])
        print("Course:", s["course"])
    else:
        print("❌ Student not found!")


def delete_student():
    roll_no = input("Enter Roll Number to delete: ")

    if roll_no in students:
        del students[roll_no]
        print("✅ Student deleted successfully!")
    else:
        print("❌ Student not found!")


def display_students():
    if not students:
        print("No students in system.")
        return

    print("\n--- All Students ---")
    for roll_no, s in students.items():
        print(f"Roll No: {roll_no}, Name: {s['name']}, Age: {s['age']}, Course: {s['course']}")


def menu():
    while True:
        print("\n===== Student Management System =====")
        print("1. Insert Student")
        print("2. Search Student")
        print("3. Delete Student")
        print("4. Display All Students")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            insert_student()
        elif choice == "2":
            search_student()
        elif choice == "3":
            delete_student()
        elif choice == "4":
            display_students()
        elif choice == "5":
            print("Exiting system...")
            break
        else:
            print("❌ Invalid choice!")


# Program Start
menu()
```
